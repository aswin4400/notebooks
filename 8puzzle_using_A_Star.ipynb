````
def solve(g, state, prev_mov):
  blank = state.index(0)
  print("Iteration ",g+1)
  h1 = 9999
  h2 = 9999
  h3 = 9999
  h4 = 9999
  if (blank - 3 >= 0 and prev_mov != 'up'):
    down = list(state)
    down[blank] = down[blank - 3]
    down[blank-3] = 0
    h1 = cal_heur(down)
    draw_board(down, 'up')
    if (h1 == 0):
      print("Thus Goal state is reached")
      return 
  if ((blank + 3) <= 8 and prev_mov != 'down'):
    up = list(state)
    up[blank] = up[blank + 3]
    up[blank + 3] = 0
    h2 = cal_heur(up)
    draw_board(up, 'down')
    if (h2 == 0):
      print("Thus Goal state is reached")
      return 
  if ((blank + 1)%3 != 0 and blank + 1 <=8 and prev_mov != 'right'):
    left = list(state)
    left[blank] = left[blank + 1]
    left[blank + 1] = 0
    h3 = cal_heur(left)
    draw_board(left, 'right')
    if (h3 == 0):
      print("Thus Goal state is reached")
      return 
  if ((blank - 1)%3 != 2 and blank - 1 >= 0 and prev_mov != 'left'):
    right = list(state)
    right[blank] = right[blank - 1]
    right[blank - 1] = 0
    h4 = cal_heur(right)
    draw_board(right, 'left')
    if (h4 == 0):
      print("Thus Goal state is reached")
      return 
  temp = 0
  move = 'null'
  if (h1 == min(h1,h2,h3,h4)):
    temp = list(down)
    mov = 'down'
  if (h2 == min(h1,h2,h3,h4)):
    temp = list(up)
    move = 'up'
  if (h3 == min(h1,h2,h3,h4)):
    temp = list(left)
    move = 'left'
  if (h4 == min(h1,h2,h3,h4)):
    temp = list(right)
    move = 'right'
  if (temp != 0 and g < 200):
    draw_board(temp, 'Least cost path')
    solve(g+1, temp, move)
  if (temp == 0 or g >=200):
    print("Unsolvable")
    return 
````
