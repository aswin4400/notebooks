# The quadratics

These are the types of equations that looks like (1)

$f(x) = x^2-5x+6\tag{1}$

The fact is these quadratic equations will form a parabolic shape. The following Ipython code displays the two ways this equation can look like. If the coefficent of $x^2$ is less that 0, then the parabola will look like the one on the right side, otherwise it will look similar to one on the left.

```python
from IPython import display
display.Image("./parabola.jpeg",width="500",height="500")
```

So if we are given with a quadratic equation, then how can we plot or describe the shape without using any tools, only pen and paper. It is apparent that the shape will be parabolic. But what will be the other parameters. Where is it located in the graph, in terms of x and y axis ? What will the deepest point of the graph ? and so on. One method, is to plot the graph by giving values to the equation and making points in the paper with reference to the two axis according to the output and joining those points to get the final output.

But this method is not recommended because of various reasons. 


1.   It is time consuming
2.   It is prone to errors
3.   It avoids all the values except, the values that we have given

Suppose we gave values 1, 2, 3 ,4 etc.. to the equation and plotted the graph based on subsequent output. But what about 1.2, 1.3, or 2.5 or -1 ? So we have ignored majority of the numbers in the number systems and only chose a few of them. So our graph can't be considered as a generic representation of that equation that incorporates all valid inputs.

To draw a parabola from a quadratic equation, We have to answer two questions.

1. Where is the **lowest point** ?
2. Where is the **Y-intercept** ?



To calculate the **Y-intercept** it is pretty much an easy process. We know **Y intercept** is the point, or the output when $x=0$. It is the $y$ when $x$ is 0. We just need to find $f(0)$ to get the Y-intercept. So if we put 0, to equation (1) it becomes, $f(0) = 6 $. So Y-intercept for the graph of equation (1) will be $+6$ on Y-axis.

The next task is to find out the Lowest possible point.

```python
display.Image('./parabola_symmetry.jpeg',width=250,height=250)
```

In ``parabola_symmetry.jpeg`` line $AB$ intercepts the parabola at some point. This point is the lowest point on the parabola. So, if we examine this plot, we can see the parabola have a symmetric shape accross the two sides of the line $AB$. This is the significance of the deepest point. It contributes very much in predicting the shape of the parabola, because of its symmetric property. Now we I understood why the deepest point is significant in determining the shape of the quadratic equation. But we still don't know how to find this point.

To find the answer let's make some modification to equation (1), without affecting the end result.

$f(x) = x^2-5x+6 + \frac{25}{4}-\frac{25}{4}\tag{2}$

equation (2) and (1) are looks different but the outputs will be same. Because we have added and subtracted $\frac{25}{4}$, which makes no change in equation (1). Now we can perform some tweaks on equation (2) form equation (3).

$f(x) = x^2-5x+\frac{25}{4}+6-\frac{25}{4}\tag{3}$

Equation (3) can also be represented as equation (4)

$f(x) = (x-\frac{5}{2})^2-\frac{1}{4}\tag{4}$

From (4) it is evident that the least possible value that $f(x)$ can have is $-\frac{1}{4}$. Because the least possible value that $(x - \frac{5}{2})^2$ can produce is 0. Thus we have found the deepest point of our parabola which is $(\frac{5}{2},-\frac{1}{4})$ and the Y-intercept $(0,+6)$. With these two points we can plot the graph of (1).
